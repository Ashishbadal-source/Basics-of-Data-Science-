```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})  #assigning no. of type of flowers 
iris.sample(5)
```

```python
# cmap is for differetn color combinations and alpha is opacity of the color and colorbar() is for that bar near the graph to show level .
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.85)    
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
plt.colorbar()
```

```python
plt.figure(figsize=(10, 6))       # to set the dimension o fthe graph ( 10 => width and 6 => height )

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.85)    
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
plt.colorbar()
```

# Annotations

```python
batter = pd.read_csv('batter.csv')
batter.head()
```

```python
sample_df = batter.head(100).sample(25,random_state=5)  # to take random 25 rows from the first 100 rows of the dataframe
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
# to describe the annotations in the graph
data_1 = [ 1,2,3,4]
data_2 = [5,6,7,8]
plt.scatter(data_1, data_2)
plt.text(1, 5, 'Point A', fontsize=12, color='red')  # Adding annotation at point (1,5)
# plt.text(x-axis wala point , y-axis wala point, jo name hme dena h usko ,uske additional features like color fontsize etc)

plt.text(2, 6, 'Point B', fontsize=10, color='green') 
plt.text(3, 7, 'Point C', fontsize=20, color='orange') 
plt.text(4, 8, 'Point D', fontdict={'fontsize':15, 'color':'blue'})  # Using a dictionary for font properties
```

```python
plt.figure(figsize=(14,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) 
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],color='red',fontsize=12)
```

```python
plt.figure(figsize=(14,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) 

plt.axhline(130,color='red',linestyle='--',label='Strike Rate Threshold')  # Adding a horizontal line at y=130
plt.axvline(30,color='blue',linestyle='--',label='Average Threshold')     #Addind a vertical line at x=25 

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],color='red',fontsize=12)
```

# Subplots

```python
batter.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batter['avg'],batter['strike_rate'],color='red')
plt.title('Batter info.')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# 2nd method for scatter plot (also use in subplot method)

plt,ax = plt.subplots(figsize=(15,6))
ax.scatter(batter['avg'],batter['strike_rate'],color='red')
ax.set_title('Batter info.')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1)
ax[0].scatter(sample_df['avg'],sample_df['strike_rate'])
ax[1].scatter(sample_df['avg'],sample_df['runs'])
plt.tight_layout()
plt.show()
```
