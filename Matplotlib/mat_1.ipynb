```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashed')
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot')
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='solid',linewidth=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2)
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dashdot',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',marker='o')
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dashdot',marker='D',markersize=8)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',marker='o',markersize=5)
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle='dashdot',marker='D',markersize=8,label='V Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',marker='o',markersize=5,label='RG Sharma')
plt.title('Comparision of runs of V Kohli and RG Sharma')
plt.xlabel('Season of the IPL')
plt.ylabel('Number of runs')

plt.legend()
```

# BAR Chart

1. Bivariate Analysis
2. Numerical vs Categorical Analysis
3. Use case : Aggregate analysis of groups

```python
children=[12,15,5,16,10,3]
colors = ['red','blue','pink','green','grey','black']

plt.bar(colors,children,color='black',edgecolor='red',width=0.5)
```

```python
children=[12,15,5,16,10,3]
colors = ['red','blue','pink','green','grey','black']

plt.barh(colors,children,color='black',edgecolor='red')
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.4)
```

```python
plt.bar(np.arange(df.shape[0])- 0.2 , df['2015'],width =0.2)
plt.bar(np.arange(df.shape[0]) , df['2016'],width =0.2)
plt.bar(np.arange(df.shape[0])+ 0.2 , df['2017'],width =0.2)
```

```python
plt.bar(np.arange(df.shape[0])- 0.2 , df['2015'],width =0.2)
plt.bar(np.arange(df.shape[0]) , df['2016'],width =0.2)
plt.bar(np.arange(df.shape[0])+ 0.2 , df['2017'],width =0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
plt.style.use('default')
```

```python
children=[12,15,5,16,10,3]
colors = ['red red color ','blue blue color ','pink pink color ','green green color ','grey grey color ','black black color ']

plt.bar(colors,children,color='black',edgecolor='red')
plt.xticks(rotation='vertical')      # if horizontal me rhta then overlap krta, so solution of that problem is this method
plt.show()
```

```python
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+ df['2016'],label='2017')

plt.legend()
plt.show()
```

# PIE Chart

1. Univariate and Bivariate Analysis
2. Numerical vs Categorical 
3. Use case : To find contribution on a standard scale

```python
data = [ 12,23,34,45,56,67,78,89]
marks = ['A','B','C','D','E','F','G','H']
plt.pie(data,labels=marks)
plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct="%.1f%%",explode=[0.3,0,0,0,0.1,0],shadow=True)
plt.show()
```
