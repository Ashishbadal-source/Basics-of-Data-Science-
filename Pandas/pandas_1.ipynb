```python
mask = data['city']=='Delhi'   # true false answer dega ye so it is called Boolean series
data[mask].shape
```

```python
def get_no_of_matches(city):
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_no_of_matches('Chennai')
```

```python
mask1 = data['city']=='Delhi'
mask2 = data['date']>'2016-04-06'   #we can now put 2,3,4.. as many condition with & and & again to achieve the req. data
data[mask1 & mask2]
```

```python
def matches_win(team):
    mask1 = data['team1']==team
    mask2 = data['team2']==team
    mask3 = data['winner']==team
    return(data[(mask1|mask2) & mask3])
```

```python
matches_win('Delhi')
```
